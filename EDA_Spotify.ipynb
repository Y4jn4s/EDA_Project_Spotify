### IMPORTING ALL FRAMEWORKS

```python
import os, warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10, 6)

DATA_FILE = 'SpotifyFeatures.csv'
```

### LOAD THE DATASET

```python
df = pd.read_csv(DATA_FILE, low_memory=False)
print("Shape:", df.shape)

display(df.head())
display(df.info())
display(df.describe(include='all').T)
```

### CLEANING AND FORMATTING

```python
df.columns = [c.strip() for c in df.columns]

nulls = df.isna().sum().sort_values(ascending=False)
print('Top null counts:')
display(nulls.head(10))

if 'popularity' in df.columns:
    df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')

if 'duration_ms' in df.columns:
    df['duration_min'] = pd.to_numeric(df['duration_ms'], errors='coerce')/60000.0

if 'release_date' in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce', infer_datetime_format=True)
    df['release_year'] = df['release_date'].dt.year
elif 'year' in df.columns:
    df['release_year'] = pd.to_numeric(df['year'], errors='coerce')

if 'track_genre' in df.columns:
    df['track_genre'] = df['track_genre'].astype(str).str.lower()

print('Finished basic cleaning steps.')
display(df[['track_name','artist_name','popularity']].head())
```

### UNIVARIATE ANALYSIS
#### AUDIO FEATURE DISTRIBUTIONS, CATEGORICAL COUNTS 

```python
num_cols = [c for c in ['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_min','popularity'] if c in df.columns]

for col in num_cols:
    plt.figure()
    df[col].dropna().hist(bins=50)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
```

```python
if 'track_genre' in df.columns:
    plt.figure(figsize=(12,5))
    df['track_genre'].value_counts().head(20).plot(kind='bar')
    plt.title('Top 20 Genres (by count)')
    plt.xlabel('Genre')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()

# Top artists
plt.figure(figsize=(12,5))
top_artists = df['artist_name'].value_counts().head(20)
top_artists.plot(kind='bar')
plt.title('Top 20 Artists (by track count)')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### POPULARITY ANALYSIS

```python
plt.figure(figsize=(8,6))
sns.histplot(df['tempo'], bins=30, kde=True, color="blue")
plt.title("Distribution of Song Tempo")
plt.xlabel("Tempo (BPM)")
plt.ylabel("Count")
plt.show()
```

```python
if 'popularity' in df.columns:
    plt.figure()
    df['popularity'].dropna().hist(bins=50)
    plt.title('Popularity distribution (0-100)')
    plt.xlabel('Popularity')
    plt.ylabel('Count')
    plt.show()

# Correlation between audio features and popularity
feature_cols = [c for c in ['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_min'] if c in df.columns]
if 'popularity' in df.columns and len(feature_cols)>0:
    corr_df = df[feature_cols + ['popularity']].corr(numeric_only=True)
    display(corr_df['popularity'].sort_values(ascending=False))
    top_pos = corr_df['popularity'].drop('popularity').abs().sort_values(ascending=False).head(4).index.tolist()
    for col in top_pos:
        plt.figure()
        plt.scatter(df[col], df['popularity'], alpha=0.3, s=10)
        plt.xlabel(col)
        plt.ylabel('popularity')
        plt.title(f'Popularity vs {col}')
        plt.tight_layout()
        plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

corr = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Spotify Features")
plt.show()
```

```python
if 'release_year' in df.columns:
    # Popularity by year
    yearly = df.groupby('release_year', as_index=False)['popularity'].mean().dropna()
    plt.plot(yearly['release_year'], yearly['popularity'], marker='o')
    plt.title('Avg Popularity by Year'); plt.xlabel('Year'); plt.ylabel('Popularity')
    plt.grid(True, ls='--', alpha=0.6); plt.show()

    # Feature trends
    feats = [f for f in ['danceability','energy','tempo','valence'] if f in df.columns]
    if feats:
        yearly_feats = df.groupby('release_year')[feats].mean()
        for f in feats:
            plt.plot(yearly_feats.index, yearly_feats[f], label=f)
        plt.title('Audio Feature Trends'); plt.xlabel('Year'); plt.ylabel('Mean Value')
        plt.legend(); plt.grid(True, ls='--', alpha=0.6); plt.show()
else:
    print(' No release_year column — skipping.')
```

### TOP HITS

```python
if 'popularity' in df.columns:
    df_unique = df.sort_values('popularity', ascending=False).drop_duplicates(
        subset=['track_name', 'artist_name'], keep='first'
    )

    top50 = df_unique.sort_values('popularity', ascending=False).head(50)
    bottom50 = df_unique.sort_values('popularity', ascending=True).head(50)

    display(top50[['track_name','artist_name','popularity'] + 
                  [c for c in feature_cols if c in top50.columns]].head(10))

    # Feature fingerprint
    fingerprint = pd.DataFrame({
        'Top 50 (mean)': top50[feature_cols].mean(),
        'Bottom 50 (mean)': bottom50[feature_cols].mean()
    })

    plt.figure(figsize=(8,6))
    sns.heatmap(fingerprint, annot=True, fmt='.2f', cmap="coolwarm", cbar=True)
    plt.title('Feature Fingerprint: Top 50 vs Bottom 50 (Unique Songs)')
    plt.show()
```

### CLUSTERING SONGS BY AUDIO FEATURES

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

cluster_cols = feature_cols
if cluster_cols:
    X = df[cluster_cols].dropna()              
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)               
    k = 4
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(Xs)
    labels = km.labels_

    # Inverse-transform cluster centers back to original feature space
    centers = pd.DataFrame(scaler.inverse_transform(km.cluster_centers_), columns=cluster_cols)
    display(centers.T)

    preview = pd.DataFrame({
        'track_name': df.loc[X.index, 'track_name'].values,
        'artist_name': df.loc[X.index, 'artist_name'].values,
        'cluster': labels
    }).reset_index(drop=True).head(200)
    print('Assigned cluster labels to first 200 songs for preview:')
    display(preview.head(10))
else:
    print('No feature columns found for clustering.')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X['energy'], X['valence'], c=labels, cmap='viridis', alpha=0.6)
plt.xlabel("Energy")
plt.ylabel("Valence")
plt.title("Clusters of Songs (Energy vs Valence)")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(Xs)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels, cmap='plasma', alpha=0.6)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Clusters of Songs (PCA Projection)")
plt.show()

#PCA PROJECTIONS
```
